## SUBSTITUINDO E PADRONIZANDO VALORES AUSENTES

```python
df_mortes_att = df_mortes.replace(['IGNORADO', 'IGNORADA', 'IGNORADO(A)', 'INDEFINIDO', 'INDEFINIDO(A)', 'INDEFINIDA', '#N/D', 'NAO INFORMADO NA GUIA DO DELEGADO', 'NI, NI, NI, NI'
], 'NI')
```

### TROCANDO VALORES NA/NAN PELO PADRÃO DEFINIDO 'NI'

```python
df_mortes_att = df_mortes_att.fillna('NI')
df_mortes_att[df_mortes_att.isna()].count()
```

### AGRUPANDO OS DADOS PARA IDENTIFICAR AS ANOMALÍAS

```python
df_mortes2.groupby(['PROFISSAO']).count()
```

### PADRONIZANDO OS VALORES DA TABELA

```python
df_mortes2_att = df_mortes2.replace(['A.S.G'], 'ASG')
df_mortes2_att = df_mortes2_att.replace(['SEM PROFISSAO'], 'DESEMPREGADO')
df_mortes2_att = df_mortes2_att.replace(['SERVENTE'], 'SERVENTE DE PEDREIRO')
df_mortes2_att.groupby(['PROFISSAO']).count()
```

### PADRONIZANDO ANOMALIAS

```python
df_mortes2_att = df_mortes2_att.replace(['NAO IDENTIFICADO'], 'NI')
```

```python
df_mortes_att.head()
```

### AJUSTANDO NOMES E ORDENS DAS COLUNAS DOS DATAFRAMES

##### RENOMEANDO COLUNAS

```python
df_mortes_att.columns = ['DATA', 'HORA', 'ANO', 'ETNIA', 'GENERO', 'IDADE', 'PROFISSAO', 'ESCOLARIDADE', 'LOGRADOURO DA OCORRENCIA', 'BAIRRO DA OCORRENCIA', 'CIDADE', 'INSTRUMENTO UTILIZADO', 'TIPO DE MORTE']
df_mortes2_att.columns = ['TIPO DE MORTE', 'DATA', 'HORA', 'ETNIA', 'GENERO', 'IDADE', 'PROFISSAO', 'ESCOLARIDADE', 'LOGRADOURO DA OCORRENCIA', 'BAIRRO DA OCORRENCIA', 'CIDADE', 'INSTRUMENTO UTILIZADO', 'ANO']
```

#### REORDENANDO AS COLUNAS

```python
df_mortes_att = df_mortes_att[['TIPO DE MORTE', 'ANO', 'DATA', 'HORA', 'CIDADE', 'BAIRRO DA OCORRENCIA', 'LOGRADOURO DA OCORRENCIA', 'GENERO', 'ETNIA', 'IDADE', 'PROFISSAO', 'ESCOLARIDADE', 'INSTRUMENTO UTILIZADO']]
df_mortes2_att = df_mortes2_att[['TIPO DE MORTE', 'ANO', 'DATA', 'HORA', 'CIDADE', 'BAIRRO DA OCORRENCIA', 'LOGRADOURO DA OCORRENCIA', 'GENERO', 'ETNIA', 'IDADE', 'PROFISSAO', 'ESCOLARIDADE', 'INSTRUMENTO UTILIZADO']]
```

#### UNINDO OS DATAFRAMES

```python
frames = [df_mortes_att, df_mortes2_att]
```

```python
df_dados = pd.concat(frames)
```
